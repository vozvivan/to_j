```python
ok = []
with open('/home/ivan/Загрузки/nedv/nor', 'r+') as all:
    for el in all:
        ok.append(el)
stud = []
with open('/home/ivan/Загрузки/nedv/stud', 'r+') as all:
    for el in all:
        stud.append(el)
print(set(ok) - set(stud))
```

```python
dict_ok = {}
for el in ok:
    try:
        dict_ok[el[:el.index('@')-1]] = el[el.index('@')+2:]
    except ValueError:
        dict_ok[el] = ''
dict_stud = {}
for el in stud:
    try:
        dict_stud[el[:el.index('@')-1]] = el[el.index('@')+2:]
    except ValueError:
        dict_stud[el] = ''
```

```python
def cmp(cat):
    try:
        print('{}\nосновной:\n{}\ncтуденеский:\n{}\n\n'.format(cat, dict_ok[cat], dict_stud[cat]))
    except KeyError:
        print('{}\nосновной:\n{}\n'.format(cat, dict_ok[cat]))
```

```python
to_cmp = ['olx.kz', 'move.su', 'olx.ua', 'realty.dmir.ru', 'domex.ru']
for el in to_cmp:
    cmp(el)
```

```python
diff = {'olx.kz':'^https?://(www\.)?([^/]+\.){0,2}olx\.kz/([^/]+/)?nedvizhimost/arenda-domov.*$', \
       'move.su':'^https?://(www\.)?([a-z_-]+\.)?move\.su/arenda_domov.*$',\
       'olx.ua':'^https?://(www\.)?([^/]+\.){0,2}olx\.ua/([^/]+/)?nedvizhimost/arenda-domov.*$',\
       'realty.dmir.ru':'^https?://(www\.)?realty\.dmir\.ru/([a-z_-]+/rent/arenda-(zagorodnoy|kottedzhey|taunhausov|domov|chastey-domov|doma)|rent/(kottedzh|taunhaus|dom|chasti-doma)).*$',\
       'domex.ru':'^https?://(www\.)?([a-z_-]+\.)?domex\.ru/((map/)?cottage/rent)|((msk/)?arenda(-posutochno)?/domov.).*$'}
```

```python
for el in to_cmp:
    dict_stud[el] = diff[el]
```

```python
with open('/home/ivan/Загрузки/nedv/all_patterns', 'w+') as all_patterns:
    for k, v in dict_stud.items():
        if v:
            all_patterns.write('{} @ {}'.format(k, v))
        else:
            all_patterns.write('{}'.format(k))
```
